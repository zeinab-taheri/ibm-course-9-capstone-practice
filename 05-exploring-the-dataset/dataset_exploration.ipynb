## Important note about `Age`

The `Age` column is not stored as normal numeric values.  
It contains text ranges such as age groups, so I cannot calculate a proper numeric mean directly from it.

To make a rough estimate, I first map each age range to one number.


```python
df["Age"].dropna().unique()[:10]
```

```python
age_map = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 70,
    "Prefer not to say": pd.NA
}

age_numeric = df["Age"].map(age_map)
estimated_mean_age = age_numeric.dropna().mean()
estimated_mean_age
```

## Why this mean is only an estimate

This is only a rough estimate because the original values are age ranges, not exact ages.
