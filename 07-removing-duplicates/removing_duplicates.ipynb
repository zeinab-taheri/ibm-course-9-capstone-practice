# Removing Duplicates with Pandas

This is my own practice notebook based on what I learned in the IBM Data Analyst Capstone Project on Coursera.

The aim is to practise basic data-cleaning steps before analysis.

## 1. Import pandas

Pandas is used to work with table-style data in Python.

```python
import pandas as pd
```

## 2. Load the dataset

The dataset is loaded from a CSV URL into a pandas DataFrame.

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)
df.head()
```

## 3. Check for duplicate rows

`duplicated()` checks whether a full row is repeated.
`sum()` counts how many duplicate rows exist.

```python
duplicate_count = df.duplicated().sum()
duplicate_count
```

## 4. View duplicate rows

This shows the duplicate rows if any exist.

```python
duplicate_rows = df[df.duplicated()]

duplicate_rows.head()
```

## 5. Remove duplicate rows

`drop_duplicates()` removes repeated rows from the DataFrame.

Important: if the duplicate count is already zero, this step does not change the dataset.

```python
df = df.drop_duplicates()

duplicates_after = df.duplicated().sum()

duplicates_after
```

## 6. Check missing values

`isnull().sum()` counts missing values in each column.

```python
missing_values = df.isnull().sum()

missing_values
```

## 7. Fill missing values in EdLevel

`EdLevel` is a categorical column, so the missing values are filled with the mode.

The mode means the most common value in the column.

```python
df["EdLevel"] = df["EdLevel"].fillna(df["EdLevel"].mode()[0])

df["EdLevel"].isnull().sum()
```

## 8. Check missing values in ConvertedCompYearly

`ConvertedCompYearly` is used for yearly compensation analysis.

It is important to check missing values before using it.

```python
df["ConvertedCompYearly"].isnull().sum()
```

## 9. Fill missing compensation values with the median

The median is the middle value.

This is a simple practice method. In real analysis, I would first investigate why many compensation values are missing.

```python
df["ConvertedCompYearly"] = df["ConvertedCompYearly"].fillna(
    df["ConvertedCompYearly"].median()
)

df["ConvertedCompYearly"].isnull().sum()
```

## 10. Final check

This confirms the dataset shape after duplicate removal and missing-value handling.

```python
df.shape
```

In this practice notebook, I checked duplicate rows, removed duplicates if present, reviewed missing values, and applied simple imputation methods using pandas.
