## Goal
Collect programming language salary data from an HTML table and save it to a CSV file.

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Source page containing the HTML table
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Download the page HTML
html_data = requests.get(url).text
html_data[:200]
```

## Parse the HTML

```python
soup = BeautifulSoup(html_data, "html.parser")
table = soup.find("table")
rows = table.find_all("tr")

len(rows)
```

## Extract the language name and average annual salary

```python
scraped_data = []

for row in rows[1:]:  # skip the header row
    cols = row.find_all("td")
    language = cols[1].get_text(strip=True)
    salary = cols[3].get_text(strip=True).replace("$", "").replace(",", "")
    scraped_data.append([language, salary])

scraped_data[:5]
```

## Save the scraped data to CSV

```python
with open("popular-languages.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Language", "Average Annual Salary"])
    writer.writerows(scraped_data)

print("Saved popular-languages.csv")
```
