## Section 6 — GET request with URL parameters

This section shows how a GET request can send extra information through the URL using query parameters.

Example parameters used here:
- `name = Joseph`
- `ID = 123`

```python
url_get='http://httpbin.org/get'
```

```python
payload={"name":"Joseph","ID":"123"}
```

```python
r=requests.get(url_get,params=payload)
```

### Review the resulting URL and request details

These cells show:
- the final URL
- the request body
- the status code
- the response text
- the response content type
- the parsed JSON arguments

```python
r.url
```

```python
print("request body:", r.request.body)
```

```python
print(r.status_code)
```

```python
print(r.text)
```

```python
r.headers['Content-Type']
```

```python
r.json()['args']
```

## Section 7 — POST request

This section shows the difference between GET and POST.

Key idea:
- GET usually sends data in the URL
- POST usually sends data in the request body

```python
url_post='http://httpbin.org/post'
```

```python
try:
    response = requests.post(url_post, data=payload)
    if response.status_code == 200:
        print("Response JSON:", response.json()) 
    else:
        print(f"HTTP Error: {response.status_code} - {response.reason}")
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
print("POST request URL:",response.url )
print("GET request URL:",r.url)
```

```python
print("POST request body:",response.request.body)
print("GET request body:",r.request.body)
```

```python
response.json()['form']
```
