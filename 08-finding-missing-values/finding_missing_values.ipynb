# Finding Missing Values with Pandas

This is my own practice notebook based on what I learned in the IBM Data Analyst Capstone Project on Coursera.

The aim is to practise identifying, counting, visualising, and handling missing values in a dataset.

## 1. Import libraries

Pandas is used for data work. Matplotlib and seaborn are used for visualisation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the dataset

The CSV file is loaded into a pandas DataFrame.

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)

df.head()
```

## 3. Explore the dataset

`info()` shows the dataset structure.

`describe()` gives summary statistics for numerical columns.

```python
df.info()
```

```python
df.describe()
```

## 4. Identify missing values across all columns

`isnull()` checks for missing values.

`sum()` counts how many missing values exist in each column.

```python
missing_values = df.isnull().sum()

missing_values
```

## 5. Visualise missing values with a heatmap

A heatmap helps show where missing values appear across the dataset.

```python
plt.figure(figsize=(20, 10))

sns.heatmap(
    df.isnull(),
    cbar=False,
    cmap="viridis"
)

plt.title("Heatmap of Missing Values", fontsize=18)
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

## 6. Count missing values in the Employment column

This checks whether the `Employment` column has missing values.

```python
missing_employment = df["Employment"].isnull().sum()

print("Number of missing rows in Employment column:", missing_employment)
```

## 7. Find the most frequent Employment value

The mode is the most common value in a column.

```python
most_frequent_employment = df["Employment"].mode()[0]

print("Most frequent value in Employment column:", most_frequent_employment)
```

## 8. Fill missing Employment values

This fills missing Employment values with the most frequent Employment value.

In the reviewed notebook, this column had zero missing values, so this step is mainly a practice/demo step.

```python
df["Employment"] = df["Employment"].fillna(most_frequent_employment)

print("Number of missing rows in Employment column:", df["Employment"].isnull().sum())
```

## 9. Visualise Employment distribution after imputation

This bar chart shows how many respondents belong to each Employment category after imputation.

```python
# Visualize the top 10 Employment categories after imputation

plt.figure(figsize=(10, 6))

df["Employment"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Employment Categories After Imputation")
plt.xlabel("Employment Type")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

## Summary

In this practice notebook, I identified missing values across the dataset, visualised missing-value patterns, checked a specific categorical column, and practised filling missing categorical values with the mode.

Important note: the `Employment` column had no missing values in the reviewed dataset, so the imputation step is a demonstration rather than a real correction.
