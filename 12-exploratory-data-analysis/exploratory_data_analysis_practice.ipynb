## Step 2: Load and preview the dataset


```python
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(data_url)

pd.set_option('display.max_columns', None)

df.head()
```

## Step 3: Handling missing data

This step checks missing values in `Employment`, `JobSat`, and `RemoteWork`, fills missing values for the lab task, and verifies the result.


```python
# Check missing values in the important columns
missing_values = df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum()
missing_percent = df[['Employment', 'JobSat', 'RemoteWork']].isnull().mean() * 100

print("Missing values:")
print(missing_values)

print("\nMissing percentage:")
print(missing_percent)

# Find the most common values
mode_jobsat = df['JobSat'].mode(dropna=True)
mode_remotework = df['RemoteWork'].mode(dropna=True)

print("\nMost common JobSat value:")
print(mode_jobsat)

print("\nMost common RemoteWork value:")
print(mode_remotework)

# Create a cleaned copy of the dataset
df_clean = df.copy()

# Drop rows where Employment is missing
df_clean = df_clean.dropna(subset=['Employment'])

# Fill missing JobSat values with the most common JobSat value
if not mode_jobsat.empty:
    df_clean['JobSat'] = df_clean['JobSat'].fillna(mode_jobsat.iloc[0])

# Fill missing RemoteWork values with the most common RemoteWork value
if not mode_remotework.empty:
    df_clean['RemoteWork'] = df_clean['RemoteWork'].fillna(mode_remotework.iloc[0])

# Check missing values again after handling them
print("\nMissing values after cleaning:")
print(df_clean[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())
```

## Step 4: Experience and job satisfaction

Important: this analysis uses real `JobSat` answers from the original dataframe, not the filled `8.0` values from `df_clean`.


```python
# Create a separate dataframe for this analysis only
experience_jobsat_df = df[['YearsCodePro', 'JobSat']].copy()

# Convert YearsCodePro into numbers properly
def convert_years(value):
    if pd.isna(value):
        return None
    elif value == 'Less than 1 year':
        return 0.5
    elif value == 'More than 50 years':
        return 50
    else:
        try:
            return float(value)
        except ValueError:
            return None

experience_jobsat_df['YearsCodePro_numeric'] = experience_jobsat_df['YearsCodePro'].apply(convert_years)

# Keep only rows with real experience and real JobSat values
experience_jobsat_df = experience_jobsat_df.dropna(subset=['YearsCodePro_numeric', 'JobSat'])

# Create experience ranges
bins = [0, 5, 10, 20, float('inf')]
labels = ['0-5', '5-10', '10-20', '>20']

experience_jobsat_df['ExperienceRange'] = pd.cut(
    experience_jobsat_df['YearsCodePro_numeric'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Calculate median JobSat for each experience range
median_jobsat_by_experience = (
    experience_jobsat_df
    .groupby('ExperienceRange', observed=False)['JobSat']
    .median()
)

print(median_jobsat_by_experience)

# Visualize the relationship
median_jobsat_by_experience.plot(kind='bar', figsize=(8, 5))

plt.title('Median Job Satisfaction by Years of Professional Coding Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction')
plt.xticks(rotation=0)
plt.show()
```

## Step 4 extra check: count, median, and mean


```python
# Check count, median, and mean for each experience group
experience_summary = experience_jobsat_df.groupby('ExperienceRange', observed=False)['JobSat'].agg(
    count='count',
    median='median',
    mean='mean'
)

experience_summary
```

## Step 5: Visualise JobSat distribution

This chart uses real `JobSat` answers only.


```python
# Use only real JobSat answers, not the filled values from df_clean
jobsat_plot_df = df.dropna(subset=['JobSat'])

plt.figure(figsize=(8, 5))

sns.countplot(
    x='JobSat',
    data=jobsat_plot_df,
    order=sorted(jobsat_plot_df['JobSat'].unique())
)

plt.title('Distribution of Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Number of Respondents')
plt.show()
```

## Step 6: Remote work preferences by employment

Important: missing `RemoteWork` values were filled with Hybrid, so Hybrid may be inflated.


```python
# Show the overall distribution of RemoteWork values
plt.figure(figsize=(8, 5))

sns.countplot(
    data=df_clean,
    x='RemoteWork',
    order=df_clean['RemoteWork'].value_counts().index
)

plt.title('Distribution of Remote Work Preferences')
plt.xlabel('Remote Work Type')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.show()

# Cross-tabulate RemoteWork preferences by Employment type
remote_work_by_employment = pd.crosstab(
    df_clean['Employment'],
    df_clean['RemoteWork']
)

remote_work_by_employment
```

## Step 7: Programming language trends by region

This shows top programming languages among survey respondents in each region.


```python
# Find the top 3 countries/regions by number of respondents
top_regions = df_clean['Country'].value_counts().head(3).index

# Create an empty dictionary to store top languages for each region
top_languages_by_region = {}

# Loop through each top region
for region in top_regions:
    
    # Filter the dataset for one region at a time
    region_df = df_clean[df_clean['Country'] == region]
    
    # Split programming languages and put each language into its own row
    languages = (
        region_df['LanguageHaveWorkedWith']
        .dropna()
        .str.split(';')
        .explode()
    )
    
    # Count the top 10 programming languages in this region
    top_languages = languages.value_counts().head(10)
    
    # Save the result
    top_languages_by_region[region] = top_languages
    
    # Print the result
    print(f"Top programming languages among survey respondents in {region}:")
    print(top_languages)
    print("\n")

    # Visualize the result with a bar chart
    top_languages.plot(kind='bar', figsize=(10, 5))
    
    plt.title(f'Top Programming Languages in {region}')
    plt.xlabel('Programming Language')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

## Step 8: Correlation between experience and satisfaction

The scatter plot visualises the relationship. The `corr()` function calculates the actual correlation value.


```python
# Create a separate dataframe for correlation analysis
correlation_df = df[['YearsCodePro', 'JobSatPoints_1']].copy()

# Convert YearsCodePro into numeric values
def convert_years(value):
    if pd.isna(value):
        return None
    elif value == 'Less than 1 year':
        return 0.5
    elif value == 'More than 50 years':
        return 50
    else:
        try:
            return float(value)
        except ValueError:
            return None

correlation_df['YearsCodePro_numeric'] = correlation_df['YearsCodePro'].apply(convert_years)

# Convert JobSatPoints_1 into numeric values
correlation_df['JobSatPoints_1'] = pd.to_numeric(
    correlation_df['JobSatPoints_1'],
    errors='coerce'
)

# Keep only rows with real values for both columns
correlation_df = correlation_df.dropna(subset=['YearsCodePro_numeric', 'JobSatPoints_1'])

# Calculate correlation
correlation_value = correlation_df['YearsCodePro_numeric'].corr(correlation_df['JobSatPoints_1'])

print("Correlation between years of experience and job satisfaction:")
print(correlation_value)

# Create scatter plot
plt.figure(figsize=(8, 5))

plt.scatter(
    correlation_df['YearsCodePro_numeric'],
    correlation_df['JobSatPoints_1'],
    alpha=0.5
)

plt.title('Correlation Between Experience and Job Satisfaction')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction Points')
plt.show()
```

## Step 9: Education and employment comparison


```python
# Create a dataframe for education and employment analysis
education_employment_df = df_clean[['EdLevel', 'Employment']].dropna()

# Create a cross-tabulation table
education_employment_table = pd.crosstab(
    education_employment_df['EdLevel'],
    education_employment_df['Employment']
)

education_employment_table
```

## Step 9 chart: Stacked bar chart

The full crosstab keeps all categories. The chart uses only the top 5 employment types to stay readable.


```python
# Use only the most common employment types to make the chart readable
top_employment_types = education_employment_df['Employment'].value_counts().head(5).index

education_employment_top = education_employment_df[
    education_employment_df['Employment'].isin(top_employment_types)
]

education_employment_chart = pd.crosstab(
    education_employment_top['EdLevel'],
    education_employment_top['Employment']
)

# Visualize education background and employment type
education_employment_chart.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6)
)

plt.title('Employment Type by Educational Background')
plt.xlabel('Educational Background')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
