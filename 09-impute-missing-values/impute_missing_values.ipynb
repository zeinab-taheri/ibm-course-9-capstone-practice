# Impute Missing Values with Pandas

This is my own practice notebook based on what I learned in the IBM Data Analyst Capstone Project on Coursera.

The aim is to practise identifying missing values and filling selected missing values using simple pandas methods.

## 1. Import pandas

Pandas is used to load, inspect, and clean table-style data.

```python
import pandas as pd
```

## 2. Load the dataset

The dataset is loaded from a CSV URL into a pandas DataFrame.

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)

df.head()
```

## 3. Check duplicate rows

Duplicate rows can distort analysis because the same record may be counted more than once.

```python
duplicate_count = df.duplicated().sum()

duplicate_count
```

## 4. Remove duplicate rows if needed

This removes repeated rows from the DataFrame.

If the duplicate count is already zero, this step does not change the dataset.

```python
df = df.drop_duplicates()

duplicates_after = df.duplicated().sum()

duplicates_after
```

## 5. Check missing values in all columns

This shows how many empty values exist in each column.

```python
missing_values = df.isnull().sum()

missing_values
```

## 6. Check missing values in RemoteWork

This focuses only on the `RemoteWork` column.

```python
remote_missing_before = df["RemoteWork"].isnull().sum()

remote_missing_before
```

## 7. Count RemoteWork categories

This shows the frequency of each work arrangement category.

```python
df["RemoteWork"].value_counts()
```

## 8. Find the most common RemoteWork value

The mode is the most frequent value in a column.

```python
most_frequent_remote = df["RemoteWork"].mode()[0]

most_frequent_remote
```

## 9. Fill missing RemoteWork values

Missing `RemoteWork` values are filled with the most common value.

This is called mode imputation.

```python
df["RemoteWork"] = df["RemoteWork"].fillna(most_frequent_remote)

df["RemoteWork"].isnull().sum()
```

## 10. Check compensation-related columns

These columns are directly related to compensation:

- `Currency`
- `CompTotal`
- `ConvertedCompYearly`

`AIComplex` is not included because it is not a compensation column.

```python
compensation_cols = ["Currency", "CompTotal", "ConvertedCompYearly"]

df[compensation_cols].describe(include="all")
```

## Summary

This notebook practises finding missing values, filling missing values in one categorical column using the mode, and checking compensation-related columns.

Important caution: filling many missing values with the most common answer is only a simple practice method. It may distort real analysis if used without further investigation.

Compensation data also needs caution because `CompTotal` contains extreme unrealistic values and `ConvertedCompYearly` still has many missing values.
